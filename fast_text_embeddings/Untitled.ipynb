```python
lang_embedding_val.sentence_embedding = lang_embedding_val.sentence_embedding.apply(lambda x: np.fromstring(x[1:-1], sep=' '))
```

```python
lookup = {l:10+i*20 for i, l in enumerate(lang_embedding_val.name.drop_duplicates().to_list())}
```

```python
def f(lang, embeddings, index):
    if index % 100000  == 0:
        print(index)
    return [f'{lookup[lang]+i}:{e}' for i, e in enumerate(embeddings)]
```
